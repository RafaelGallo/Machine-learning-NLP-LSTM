```python
# Minimo e maximo dos preÃ§os

df_max = df.price.max()
df_min = df.price.min()

print("MÃ¡ximo price", df_max)
print("Minimo price", df_min)
```

```python
# MÃ©dia, mediana, moda

price_media = df.price.mean()
print("MÃ©dia preÃ§os dos imÃ³veis:", price_media)

price_mediana = df.price.median()
print("MÃ©diana preÃ§os dos imÃ³veis:", price_mediana)

price_mod = df.price.mode()
print("Moda preÃ§os dos imÃ³veis:", price_mod)
```

```python
# Matriz correlaÃ§Ã£o de pares de colunas, excluindo NA / valores nulos.

corr = df.corr()
corr
```

```python
# GrÃ¡fico da matriz de correlaÃ§Ã£o

plt.figure(figsize=(35.5,30))
ax = sns.heatmap(corr, annot=True, cmap='YlGnBu');
plt.title("Matriz de correlaÃ§Ã£o")
```

```python
# Verificando os dados no boxplot valor total verificando possÃ­veis outliers

plt.figure(figsize=(20.5, 18))

sns.boxplot(x="price", y="room_type", data = df)
plt.title("PreÃ§os dos imÃ³veis por tipo")
plt.xlabel("ImÃ³veis e os tipos")
plt.ylabel("Total")
```

# **AnÃ¡lise dados**

```python
# GrÃ¡fico preÃ§os dos imÃ³veis pelo bairro
plt.figure(figsize=(18.5, 8))

plt.title("PreÃ§os dos imÃ³veis por bairro")
sns.barplot(x="neighbourhood_group", y="price", data = df)
plt.xlabel("Bairros de Nova York")
plt.ylabel("Total")
```

```python
# Visualizando grÃ¡fico pÃ³ssivil outliers na base de price

plt.figure(figsize=(18.5, 8))
sns.boxplot(df["price"])
plt.title("GrÃ¡fico de boxplot dos preÃ§os dos imÃ³veis")
plt.xlabel("PreÃ§os dos imÃ³veis")
plt.ylabel("Total")
```

```python
# GrÃ¡fico histrograma de preÃ§os dos imÃ³veis
plt.figure(figsize=(18.5, 8))

sns.histplot(df["price"])
plt.title("GrÃ¡fico dpreÃ§os dos imÃ³veis")
plt.xlabel("PreÃ§os dos imÃ³veis")
plt.ylabel("Total")
```

```python
# GrÃ¡fico de barras tipos de aparatamentos e casas 
plt.figure(figsize=(18.5, 8))

sns.countplot(df["room_type"])
plt.title("Tipos de casas NY")
plt.xlabel("Bairros de Nova York")
plt.ylabel("Total")
```

```python
# GrÃ¡fico de boxplot vizinhanÃ§a e avaliaÃ§Ã£o dos imÃ³veis
plt.figure(figsize=(18, 8))

sns.boxplot(x = "neighbourhood_group", y = "availability_365", data = df)
plt.title("Bairros de NY pela avaliaÃ§Ã£o")
plt.xlabel("Bairros NY")
plt.ylabel("Total")
```

```python
# GrÃ¡fico scatterplot mapa da regiÃ£o de NY
plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "neighbourhood_group")

plt.title("Mapa dos imÃ³veis - Nova York NY")
plt.xlabel("Bairros NY")
plt.ylabel("Total")
```

```python
# GrÃ¡fico de scatterplot mapa Nova York e tipos de imÃ³veis

plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "room_type")

plt.title("GrÃ¡fico mapa Nova York - Tipos apartamentos")
plt.xlabel("Mapa de Nova york")
```

```python
# GrÃ¡fico de scatterplot - Mapa de nova york por preÃ§os dos imÃ³veis

plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "price")
plt.title("Mapa Nova York por preÃ§o dos imÃ³veis")
plt.xlabel("Mapa NY")
plt.ylabel("Total")
```
