```python
# Importando as bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Total de colunas e linhas - data_train

print("Números de linhas: {}" .format(df.shape[0]))
print("Números de colunas: {}" .format(df.shape[1]))
```

```python
# Exibindo valores ausentes e valores únicos

print("\nMissing values :  ", df.isnull().sum().values.sum())
print("\nUnique values :  \n",df.nunique())
```

```python
# Dados variaveis
df.var()
```

```python
# Minimo e maximo dos preços

df_max = df.price.max()
df_min = df.price.min()

print("Máximo price", df_max)
print("Minimo price", df_min)
```

```python
# Média, mediana, moda

price_media = df.price.mean()
print("Média preços dos imóveis:", price_media)

price_mediana = df.price.median()
print("Médiana preços dos imóveis:", price_mediana)

price_mod = df.price.mode()
print("Moda preços dos imóveis:", price_mod)
```

```python
# Exibindo estatísticas descritivas visualizar alguns detalhes estatísticos básicos como percentil, média, padrão, etc. 

# De um quadro de dados ou uma série de valores numéricos.

df.describe().T
```

```python
df.describe()
```

```python
# Matriz correlação de pares de colunas, excluindo NA / valores nulos.

corr = df.corr()
corr
```

```python
# Gráfico da matriz de correlação

plt.figure(figsize=(35.5,30))
ax = sns.heatmap(corr, annot=True, cmap='YlGnBu');
plt.title("Matriz de correlação")
```

```python
# Verificando os dados no boxplot valor total verificando possíveis outliers

plt.figure(figsize=(20.5, 18))

sns.boxplot(x="price", y="room_type", data = df)
plt.title("Preços dos imóveis por tipo")
plt.xlabel("Imóveis e os tipos")
plt.ylabel("Total")
```

# **Análise dados**

```python
# Gráfico preços dos imóveis pelo bairro
plt.figure(figsize=(18.5, 8))

plt.title("Preços dos imóveis por bairro")
sns.barplot(x="neighbourhood_group", y="price", data = df)
plt.xlabel("Bairros de Nova York")
plt.ylabel("Total")
```

```python
# Visualizando gráfico póssivil outliers na base de price

plt.figure(figsize=(18.5, 8))
sns.boxplot(df["price"])
plt.title("Gráfico de boxplot dos preços dos imóveis")
plt.xlabel("Preços dos imóveis")
plt.ylabel("Total")
```

```python
# Gráfico histrograma de preços dos imóveis
plt.figure(figsize=(18.5, 8))

sns.histplot(df["price"])
plt.title("Gráfico dpreços dos imóveis")
plt.xlabel("Preços dos imóveis")
plt.ylabel("Total")
```

```python
# Gráfico de barras tipos de aparatamentos e casas 
plt.figure(figsize=(18.5, 8))

sns.countplot(df["room_type"])
plt.title("Tipos de casas NY")
plt.xlabel("Bairros de Nova York")
plt.ylabel("Total")
```

```python
# Gráfico de boxplot vizinhança e avaliação dos imóveis
plt.figure(figsize=(18, 8))

sns.boxplot(x = "neighbourhood_group", y = "availability_365", data = df)
plt.title("Bairros de NY pela avaliação")
plt.xlabel("Bairros NY")
plt.ylabel("Total")
```

```python
# Gráfico scatterplot mapa da região de NY
plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "neighbourhood_group")

plt.title("Mapa dos imóveis - Nova York NY")
plt.xlabel("Bairros NY")
plt.ylabel("Total")
```

```python
# Gráfico de scatterplot mapa Nova York e tipos de imóveis

plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "room_type")

plt.title("Gráfico mapa Nova York - Tipos apartamentos")
plt.xlabel("Mapa de Nova york")
```

```python
# Gráfico de scatterplot - Mapa de nova york por preços dos imóveis

plt.figure(figsize=(10.8, 16))

sns.scatterplot(x = "latitude", 
                y = "longitude", 
                data = df, 
                hue = "price")
plt.title("Mapa Nova York por preço dos imóveis")
plt.xlabel("Mapa NY")
plt.ylabel("Total")
```
